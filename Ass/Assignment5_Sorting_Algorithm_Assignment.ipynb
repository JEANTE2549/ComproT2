## Bubble Sort 

```python
def bubbleSort(seq):
    n = len(seq) - 1
    for i in range(n,0,-1):
        for j in range(i):
            if seq[j] > seq[j+1]:
                seq[j], seq[j+1] = seq[j+1], seq[j]
    return seq
```

## Selection Sort 

```python
def selectionSort(seq):
    n = len(seq)
    for i in range(n-1):
        smallNdex = i
        for j in range(i+1, n):
            if seq[j] < seq[smallNdex]:
                smallNdex = j
        if smallNdex != i:
            seq[i], seq[smallNdex] = seq[smallNdex], seq[i]
```

## Insertion Sort

```python
def insertionSort(seq):
    n = len(seq)
    for i in range(1, n):
        value = seq[i]
        pos = i
        while pos > 0 and value < seq[pos - 1]:
            seq[pos] = seq[pos - 1]
            pos -= 1
        seq[pos] = value
    return seq
```

## Experiment1: Compare sorting algorithm in Average Case

```python
from Lib import Array
def duplicate2Array(source):
    n = len(source)
    destination = Array(n)
    for i in range(n):
        destination[i] = source[i]
    return destination
```

```python
# Prepare Sequence
n = 10
random.seed(0)
list_seq = [random.randint(0,n) for x in range(n)] # <-- Avg case
#list_seq.sort() # Best case
#list_seq.sort(reverse=True) # Worst case
array_seq = duplicate2Array(list_seq)
print(array_seq)
```

```python
# Measure execute time of function
import time
def measure_time(func,param):
    arr = duplicate2Array(param)
    start = time.perf_counter()
    func(arr)
    end = time.perf_counter()
    return end - start
```

```python
def timeExecution(n, case = "average"):
    random.seed(0)
    print(f'n = {n}')
    list_seq = [random.randint(0,n) for x in range(n)]
    if case == "best":
        list_seq.sort()
    elif case == "worst":
        list_seq.sort(reverse=True)
    elapsed_time_bubble = measure_time(bubbleSort,list_seq)
    elapsed_time_select = measure_time(selectionSort,list_seq)
    elapsed_time_insert = measure_time(insertionSort,list_seq)
    print(elapsed_time_bubble)
    print(elapsed_time_select)
    print(elapsed_time_insert)
    print()
    return [elapsed_time_bubble, elapsed_time_select, elapsed_time_insert]
```

```python
size = [1000, 2000, 5000, 10000, 15000, 20000]
bubbleTime = []
selectTime = []
insertTime = []

for i in size:
    tE = timeExecution(i)
    bubbleTime.append(tE[0])
    selectTime.append(tE[1])
    insertTime.append(tE[2])
```

```python
plt.figure()
plt.plot(size, bubbleTime, marker="o", label="Bubble Sort")
plt.plot(size, selectTime, marker="s", label="Selection Sort")
plt.plot(size, insertTime, marker="^", label="Insertion Sort")

plt.title("Execution Time Comparison of Sorting Algorithms")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
size = [1000, 2000, 5000, 10000, 15000, 20000]
bubbleTime = []
selectTime = []
insertTime = []

for i in size:
    tE = timeExecution(i, "best")
    bubbleTime.append(tE[0])
    selectTime.append(tE[1])
    insertTime.append(tE[2])
```

## Discussion

The graphs beautifully illustrate the physical memory operations of these algorithms in Python. In the best-case scenario, Insertion Sort flattens out to a lightning-fast $O(n)$ execution, while Bubble Sort and Selection Sort perfectly overlap, confirming that your unoptimized Bubble Sort is still making all $O(n^2)$ comparisons without an early-exit condition. However, in the worst-case scenario with a reverse-sorted array, Selection Sort becomes the undeniable winner; by strictly limiting its memory writes to $O(n)$ swaps, it easily outperforms Bubble and Insertion Sorts, which get completely bogged down by $O(n^2)$ constant swapping and shifting operations.
