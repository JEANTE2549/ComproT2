## Bubble Sort 

```python
def bubbleSort(seq):
    n = len(seq) - 1
    for i in range(n,0,-1):
        for j in range(i):
            if seq[j] > seq[j+1]:
                seq[j], seq[j+1] = seq[j+1], seq[j]
                print(seq) #print output ทุก iterate
    return seq
```

## Selection Sort 

```python
def selectionSort(seq):
    n = len(seq)
    for i in range(n-1):
        smallNdx = i
        for j in range(i+1, n):
            if seq[j] < seq[smallNdx]:
                smallNdx = j
        if smallNdx != i:
            seq[i], seq[smallNdx] = seq[smallNdx], seq[i]
```

> test and learn selection sort algorithm

```python
# Create a random list of number from 1 - 1000, n numbers 
import random
n = 10
seq_C = [random.randint(1,100) for x in range(n)]
print(seq_C)
```

```python
seq_D = selectionSort(seq_C)
print(seq_D)
```

## Insertion Sort

```python
def insertionSort(seq):
    n = len(seq)
    for i in range(1, n):
        value = seq[i]
        pos = i
        while pos > 0 and value < seq[pos - 1]:
            seq[pos] = seq[pos - 1]
            pos -= 1
        seq[pos] = value
    return seq
```

> test and learn insertion sort algorithm

```python
# Create a random list of number from 1 - 1000, n numbers 
import random
n = 10
seq_E = [random.randint(1,100) for x in range(n)]
print(seq_E)
```

```python
seq_F = insertionSort(seq_E)
print(seq_F)
```

### Test when we sort with completely create a backup sequence (= one by one)

```python
def duplicate(source,destination):
    n = len(source)
    for i in range(0,n):
        destination[i] = source[i]
    return destination
```
