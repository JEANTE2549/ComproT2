```python
class Updatematrix(Matrix):

    def det(self):
        assert ((self.nRows == 2 and self.nCols == 2) or (self.nRows == 3 and self.nCols == 3)), "ไม่สามารถคำนวนได้"
        if self.nRows == 2 and self.nCols == 2:
            a = self._theRows[0][0]
            b = self._theRows[0][1]
            c = self._theRows[1][0]
            d = self._theRows[1][1]
            return (a*d) - (b*c)

        elif self.nRows == 3 and self.nCols == 3:
            a = self._theRows[0][0]
            b = self._theRows[0][1]
            c = self._theRows[0][2]
            d = self._theRows[1][0]
            e = self._theRows[1][1]
            f = self._theRows[1][2]
            g = self._theRows[2][0]
            h = self._theRows[2][1]
            i = self._theRows[2][2]
            return (a*e*i) + (b*f*g) + (c*d*h) - (c*e*g) - (b*d*i) - (a*f*h)
    
    def det2(self, A):
        a = A[0][0]
        b = A[0][1]
        c = A[0][2]
        d = A[1][0]
        e = A[1][1]
        f = A[1][2]
        g = A[2][0]
        h = A[2][1]
        i = A[2][2]
        return (a*e*i) + (b*f*g) + (c*d*h) - (c*e*g) - (b*d*i) - (a*f*h)
        
    def cramer(self, B):
        if self.det() == 0:
            return "ไม่มีคำตอบเฉพาะ"
        else:
            print(self.D())
            print(self.Dx(B))
            print(self.Dy(B))
            print(self.Dz(B))
            print(self.det2(self.D()))
            print(self.det2(self.Dx(B)))
            print(self.det2(self.Dy(B)))
            print(self.det2(self.Dz(B)))
            x = self.det2(self.Dx(B)) / self.det2(self.D())
            y = self.det2(self.Dy(B)) / self.det2(self.D())
            z = self.det2(self.Dz(B)) / self.det2(self.D())
            return f"x = {x}, y= {y}, z = {z}"

    def D(self):
        a = self._theRows[0][0]
        b = self._theRows[0][1]
        c = self._theRows[0][2]
        d = self._theRows[1][0]
        e = self._theRows[1][1]
        f = self._theRows[1][2]
        g = self._theRows[2][0]
        h = self._theRows[2][1]
        i = self._theRows[2][2]
        return [[a,b,c],[d,e,f],[g,h,i]]

    def Dx(self, B):
        a = B[0]
        b = self._theRows[0][1]
        c = self._theRows[0][2]
        d = B[1]
        e = self._theRows[1][1]
        f = self._theRows[1][2]
        g = B[2]
        h = self._theRows[2][1]
        i = self._theRows[2][2]
        return [[a,b,c],[d,e,f],[g,h,i]]
    
    def Dy(self, B):
        a = self._theRows[0][0]
        b = B[0]
        c = self._theRows[0][2]
        d = self._theRows[1][0]
        e = B[1]
        f = self._theRows[1][2]
        g = self._theRows[2][0]
        h = B[2]
        i = self._theRows[2][2]
        return [[a,b,c],[d,e,f],[g,h,i]]

    def Dz(self, B):
            a = self._theRows[0][0]
            b = self._theRows[0][1]
            c = B[0]
            d = self._theRows[1][0]
            e = self._theRows[1][1]
            f = B[1]
            g = self._theRows[2][0]
            h = self._theRows[2][1]
            i = B[2]
            return [[a,b,c],[d,e,f],[g,h,i]]
```
