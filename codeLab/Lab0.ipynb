## OOP (Object Oriented Programming)

### What is an **Object?**

Each object is an instance of a class

### Design Principle

**Modularity** Modularity is about independence. A modular system is built of distinct parts that can be built, tested, and modified separately.

**Module** A file or container that groups related code (functions, classes, variables) together to organize the namespace and manage scope.

**Abstraction** Hiding complex implementation details and showing only the necessary features of an object to the user.

**Encapsulation** Bundling data (attributes) and methods (functions) into a single unit (class) and restricting direct access to some of an object's components.

## Inheritance

````python
class Dog: # Class installation
    def __init__(self, color, eye, height, lenght, weight): # Properties / Parameter
        self._color = color # Attribute
        self._eye = eye
        self._height = height
        self._lenght = lenght
        self._weight = weight
    def sit(self): # methods
        return "sit"
    def layDown(self):
        return "lay down"
    def shake(self):
        return "shake"
    def come(self):
        return "come"
````

````python
# Create object
D = Dog("Blue", "Black", 50, 50, 50 ) # send agrument back to parameter
````

````python
class Complex:
    def __init__(self, realpart, imgpart):
        self.r = realpart
        self.i = imgpart
````

````python
class modifier:
    def __init__(self, pri, pub, pro):
        self._protect = pro # Protect
        self.__private = pri # Private
        self.public = pub # Public
````

````python
class A(object):
    def __method(self):
        print('I am a method in A')
    def method(self): # getter
        self.__method()

class B(A):
    def __method(self):
        print('I am a method in B')
````
