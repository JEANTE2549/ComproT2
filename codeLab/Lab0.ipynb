## OOP (Object Oriented Programming)

### What is an **Object?**

Each object is an instance of a class

### Design Principle

**Modularity** Modularity is about independence. A modular system is built of distinct parts that can be built, tested, and modified separately.

**Module** A file or container that groups related code (functions, classes, variables) together to organize the namespace and manage scope.

**Abstraction** Hiding complex implementation details and showing only the necessary features of an object to the user.

**Encapsulation** Bundling data (attributes) and methods (functions) into a single unit (class) and restricting direct access to some of an object's components.
