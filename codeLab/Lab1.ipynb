# Array ADT

```python
import ctypes

class Array:
    
    def __init__(self,size):
        self._size = size
        arrayType = ctypes.py_object * size
        self.slots = arrayType()
        self.clear(None)

    def __len__(self):
        return self._size
    
    def __getitem__(self, pos):
        print(self.slots[pos])

    def __setitem__(self, pos, val):
        self.slots[pos] = val

    def clear (self, val):
        for i in range(len(self)):
            self.slots[i] = val
```

```python
A = Array(5)

A.__setitem__(0,1)
A.__getitem__(2)
```
